### Import libraries

```python
import os
import pandas as pd
```

### Reading the opinions files and inserting them into a combined file

```python
folder_path = "./opinions"

dfs = []

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        full_path = os.path.join(folder_path, filename)
        df = pd.read_csv(full_path)
        dfs.append(df)

# Combine all CSVs
combined_df = pd.concat(dfs, ignore_index=True)
combined_df = combined_df.fillna("no data")

# Enforce the exact column order
required_cols = [
    "opinion_id",
    "product_name",
    "sentiment",
    "opinion",
    "score",
    "date"
]
combined_df = combined_df[required_cols]

combined_df.to_csv("consolidated_opinions.csv", index=False)
combined_df
```
